```python
# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense, Dropout
from tensorflow.keras.callbacks import Callback

# ë°ì´í„° ë¡œë“œ
file_path = 'ai-ga-dataset.csv'
df = pd.read_csv("C:\\Users\\ì…œë¡\\Downloads\\ai-ga-dataset.csv")

# ë°ì´í„° í™•ì¸
print("--- ë°ì´í„°ì…‹ ìƒìœ„ 5ê°œ í–‰ ---")
print(df.head())
print("\n--- ë°ì´í„°ì…‹ ì •ë³´ ---")
print(df.info())

# 'title'ê³¼ 'abstract'ë¥¼ ê²°í•©í•˜ì—¬ í…ìŠ¤íŠ¸ ë°ì´í„° ìƒì„±
# NaN ê°’ì€ ë¹ˆ ë¬¸ìì—´ë¡œ ëŒ€ì²´
df['text'] = df['title'].fillna('') + ' ' + df['abstract'].fillna('')
X = df['text'].values
y = df['label'].values
```

```python
# 2-1. ë°ì´í„° ë¶„í•  (7:1:2 = í›ˆë ¨:ê²€ì¦:í…ŒìŠ¤íŠ¸)
# ì´ ë°ì´í„°ì˜ 20%ë¥¼ ìµœì¢… í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ë¶„í• 
test_size_ratio = 0.2
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=test_size_ratio, random_state=42, stratify=y)

# ë‚¨ì€ ë°ì´í„°(80%)ì—ì„œ ê²€ì¦ ë°ì´í„°(ì´ ë°ì´í„°ì˜ 10% = 80%ì˜ 12.5%)ë¥¼ ë¶„í• 
# 10 / 80 = 0.125
validation_size_ratio_of_temp = 0.125
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=validation_size_ratio_of_temp, random_state=42, stratify=y_temp)

print(f"**ë°ì´í„° ë¶„í•  ê²°ê³¼ (7:1:2)**")
print(f"í›ˆë ¨ ë°ì´í„° ìƒ˜í”Œ ìˆ˜: {len(X_train)} (70%)")
print(f"ê²€ì¦ ë°ì´í„° ìƒ˜í”Œ ìˆ˜: {len(X_val)} (10%)")
print(f"í…ŒìŠ¤íŠ¸ ë°ì´í„° ìƒ˜í”Œ ìˆ˜: {len(X_test)} (20%)")
print("-" * 30)

# 2-2. í…ìŠ¤íŠ¸ ì „ì²˜ë¦¬ ë§¤ê°œë³€ìˆ˜ ì„¤ì •
vocab_size = 10000    # ì‚¬ìš©í•  ë‹¨ì–´ì˜ ìµœëŒ€ ê°œìˆ˜
embedding_dim = 100   # ì„ë² ë”© ë²¡í„° ì°¨ì›
maxlen = 200          # ë¬¸ì¥ì˜ ìµœëŒ€ ê¸¸ì´ (ì´ë³´ë‹¤ ê¸´ ë¬¸ì¥ì€ ì˜ë¦¬ê³ , ì§§ì€ ë¬¸ì¥ì€ ì±„ì›Œì§)

# í† í°í™” (ë‹¨ì–´ë¥¼ ìˆ«ìë¡œ ë§¤í•‘)
tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train) # í›ˆë ¨ ë°ì´í„°ì—ë§Œ fit!

# í…ìŠ¤íŠ¸ë¥¼ ì •ìˆ˜ ì‹œí€€ìŠ¤ë¡œ ë³€í™˜
X_train_sequences = tokenizer.texts_to_sequences(X_train)
X_val_sequences = tokenizer.texts_to_sequences(X_val) # ê²€ì¦ ë°ì´í„° ì¶”ê°€
X_test_sequences = tokenizer.texts_to_sequences(X_test)

# ì‹œí€€ìŠ¤ë¥¼ ìµœëŒ€ ê¸¸ì´ë¡œ íŒ¨ë”©
X_train_padded = pad_sequences(X_train_sequences, maxlen=maxlen, padding='post', truncating='post')
X_val_padded = pad_sequences(X_val_sequences, maxlen=maxlen, padding='post', truncating='post') # ê²€ì¦ ë°ì´í„° ì¶”ê°€
X_test_padded = pad_sequences(X_test_sequences, maxlen=maxlen, padding='post', truncating='post')

print(f"í›ˆë ¨ ë°ì´í„° íŒ¨ë”© í›„ shape: {X_train_padded.shape}")
print(f"ê²€ì¦ ë°ì´í„° íŒ¨ë”© í›„ shape: {X_val_padded.shape}")
print(f"í…ŒìŠ¤íŠ¸ ë°ì´í„° íŒ¨ë”© í›„ shape: {X_test_padded.shape}")
```

```python
# 3. 1D CNN ëª¨ë¸ êµ¬ì¶•
model = Sequential([
    # í…ìŠ¤íŠ¸ë¥¼ ë²¡í„°ë¡œ ë³€í™˜
    Embedding(vocab_size, embedding_dim, input_length=maxlen),
    
    # 1D ì»¨ë³¼ë£¨ì…˜ ë ˆì´ì–´ (í…ìŠ¤íŠ¸ì—ì„œ íŠ¹ì§• ì¶”ì¶œ)
    Conv1D(filters=128, kernel_size=5, activation='relu'),
    
    # íŠ¹ì§• ë§µì—ì„œ ê°€ì¥ ì¤‘ìš”í•œ íŠ¹ì§• ì„ íƒ
    GlobalMaxPooling1D(),
    
    # ê³¼ì í•© ë°©ì§€
    Dropout(0.5),
    
    # ìµœì¢… ë¶„ë¥˜ë¥¼ ìœ„í•œ ë°€ì§‘ ì—°ê²° ë ˆì´ì–´
    Dense(64, activation='relu'),
    
    # ì´ì§„ ë¶„ë¥˜ë¥¼ ìœ„í•œ ì¶œë ¥ ë ˆì´ì–´ (sigmoid)
    Dense(1, activation='sigmoid')
])

# ëª¨ë¸ ì»´íŒŒì¼
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

print("\n--- ëª¨ë¸ êµ¬ì¡° ìš”ì•½ ---")
model.summary()
print("-" * 30)

# 4. ì‚¬ìš©ì ì •ì˜ ì½œë°± ì •ì˜ 
class EpochPerformanceCallback(Callback):
    def __init__(self, X_test, y_test, target_epochs=[3, 5, 10]):
        super().__init__()
        
        self.X_test = X_test
        self.y_test = y_test
        self.target_epochs = target_epochs
        self.results = []  # ê²°ê³¼ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸

    def on_epoch_end(self, epoch, logs=None):
        # ì—í¬í¬ ë²ˆí˜¸ëŠ” 0ë¶€í„° ì‹œì‘í•˜ë¯€ë¡œ 1ì„ ë”í•©ë‹ˆë‹¤.
        current_epoch = epoch + 1
        
        if current_epoch in self.target_epochs:
            print(f"\n\n--- Epoch {current_epoch} ì„±ëŠ¥ í‰ê°€ (í…ŒìŠ¤íŠ¸ ë°ì´í„°) ---")
            
            # 1. ì˜ˆì¸¡ ìˆ˜í–‰
            y_pred_proba = self.model.predict(self.X_test, verbose=0)
            # 0.5ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì´ì§„ ë¶„ë¥˜ ê²°ê³¼ (0 ë˜ëŠ” 1)ë¡œ ë³€í™˜
            y_pred = (y_pred_proba > 0.5).astype(int).flatten()

            # 2. í˜¼ë™ í–‰ë ¬ ê³„ì‚°
            cm = confusion_matrix(self.y_test, y_pred)
            
            # í˜¼ë™ í–‰ë ¬ì´ 2x2ì¸ì§€ í™•ì¸
            if cm.shape == (2, 2):
                tn, fp, fn, tp = cm.ravel()
            else:
                # í•œìª½ í´ë˜ìŠ¤ë§Œ ì˜ˆì¸¡ëœ ê²½ìš° ì²˜ë¦¬
                print(f"í˜¼ë™ í–‰ë ¬ì´ 2x2ê°€ ì•„ë‹™ë‹ˆë‹¤. ì‹¤ì œ shape: {cm.shape}")
                print(f"í˜¼ë™ í–‰ë ¬:\n{cm}")
                # ê¸°ë³¸ê°’ ì„¤ì •
                tn = fp = fn = tp = 0
                if cm.shape[0] == 1 and cm.shape[1] == 1:
                    # ëª¨ë“  ì˜ˆì¸¡ì´ ê°™ì€ í´ë˜ìŠ¤ì¸ ê²½ìš°
                    if self.y_test[0] == y_pred[0]:
                        if y_pred[0] == 0:
                            tn = cm[0, 0]
                        else:
                            tp = cm[0, 0]
            
            print(f"**[í…ŒìŠ¤íŠ¸ ë°ì´í„° í˜¼ë™ í–‰ë ¬ ì§€í‘œ]**")
            print(f"1. ì›ë˜ ë ˆì´ë¸” 0ì¸ë° 0ìœ¼ë¡œ ì˜ˆì¸¡í•œ ê²ƒ (TN): {tn}")
            print(f"2. ì›ë˜ ë ˆì´ë¸” 0ì¸ë° 1ìœ¼ë¡œ ì˜ˆì¸¡í•œ ê²ƒ (FP): {fp}")
            print(f"3. ì›ë˜ ë ˆì´ë¸” 1ì¸ë° 0ìœ¼ë¡œ ì˜ˆì¸¡í•œ ê²ƒ (FN): {fn}")
            print(f"4. ì›ë˜ ë ˆì´ë¸” 1ì¸ë° 1ìœ¼ë¡œ ì˜ˆì¸¡í•œ ê²ƒ (TP): {tp}")

            # ê²°ê³¼ë¥¼ ë¦¬ìŠ¤íŠ¸ì— ì €ì¥
            self.results.append({
                'Epoch': current_epoch,
                'TN': tn,
                'FP': fp,
                'FN': fn,
                'TP': tp
            })
            print("-" * 50)

    def on_train_end(self, logs=None):
        # í•™ìŠµì´ ëë‚œ í›„ ê²°ê³¼ë¥¼ í‘œë¡œ ì¶œë ¥
        self.print_summary_table()

    def print_summary_table(self):
        if not self.results:
            return

        print("\n\n## ğŸ“‹ ì—í¬í¬ë³„ í…ŒìŠ¤íŠ¸ ë°ì´í„° ì„±ëŠ¥ í‰ê°€ ìš”ì•½ (7:1:2)")
        print("-" * 75)
        print("| ì—í¬í¬ | TN (0->0) | FP (0->1) | FN (1->0) | TP (1->1) |")
        print("|--------|-----------|-----------|-----------|-----------|")
        
        for r in self.results:
            print(f"| {r['Epoch']:6} | {r['TN']:9} | {r['FP']:9} | {r['FN']:9} | {r['TP']:9} |")
        print("-" * 75)

# 5. ëª¨ë¸ í•™ìŠµ ë° ì½œë°± ì ìš©
performance_callback = EpochPerformanceCallback(X_test_padded, y_test)

history = model.fit(X_train_padded, y_train,
                    epochs=10,  # ì—í¬í¬ ìˆ˜: í•™ìŠµ íšŸìˆ˜
                    batch_size=32,
                    validation_data=(X_val_padded, y_val),  # ê²€ì¦ ë°ì´í„° ì‚¬ìš©
                    callbacks=[performance_callback],  # ì½œë°± ì¶”ê°€!
                    verbose=1)

print("\n**ëª¨ë¸ í•™ìŠµ ì™„ë£Œ!**")
```

```python
# ìƒˆ ì…€: ëª¨ë¸ ì„¤ëª… ë° ë¹Œë“œ ê²°ê³¼ ìš”ì•½ í™•ì¸

print("="*50)
print(" 1D CNN ëª¨ë¸ ìš”ì•½ ë° íŒŒë¼ë¯¸í„° í™•ì¸ (ë¹Œë“œ ì™„ë£Œ í›„)")
print("="*50)

# 1. ë¹Œë“œëœ ëª¨ë¸ êµ¬ì¡° ì¶œë ¥ (íŒŒë¼ë¯¸í„° ìˆ˜ í™•ì •)
# ì´ ì‹œì ì—ì„œëŠ” model.fit()ì´ ì‹¤í–‰ë˜ì—ˆê¸° ë•Œë¬¸ì— '?'ì™€ '0 (unbuilt)'ì´ ì‚¬ë¼ì§„ ìƒíƒœì…ë‹ˆë‹¤.
model.summary()

print("\n--- íŒŒë¼ë¯¸í„° ì„¤ëª… ---")
print(f"**Total params (ì´ ê°€ì¤‘ì¹˜):** ëª¨ë¸ ì „ì²´ì˜ ëª¨ë“  í•™ìŠµ ê°€ëŠ¥í•œ/ë¶ˆê°€ëŠ¥í•œ ê°€ì¤‘ì¹˜ ê°œìˆ˜ì…ë‹ˆë‹¤.")

# 2. Embedding ë ˆì´ì–´ íŒŒë¼ë¯¸í„° ê³„ì‚° ì„¤ëª… ì¶”ê°€
# (vocab_size * embedding_dim)
vocab_size = 10000 
embedding_dim = 100
embedding_params = vocab_size * embedding_dim

print(f"**Embedding Layer Param #**:")
print(f"  - ê³„ì‚°: ë‹¨ì–´ ìˆ˜({vocab_size}) x ì„ë² ë”© ì°¨ì›({embedding_dim}) = {embedding_params:,}ê°œ")
print(f"  - ì—­í• : í…ìŠ¤íŠ¸ì˜ ê° ë‹¨ì–´ë¥¼ 100ì°¨ì›ì˜ ì—°ì†ì ì¸ ë²¡í„°ë¡œ í‘œí˜„í•˜ëŠ” ê°€ì¤‘ì¹˜ì…ë‹ˆë‹¤.")
```

```python
# 4-1. í•™ìŠµ ê³¡ì„  ì‹œê°í™”
plt.figure(figsize=(14, 6))

# ì •í™•ë„ (Accuracy) ì‹œê°í™”
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy', fontsize=15)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.legend()
plt.grid(True)

# ì†ì‹¤ (Loss) ì‹œê°í™”
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss', fontsize=15)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# 4-2. í…ŒìŠ¤íŠ¸ ë°ì´í„° ìµœì¢… í‰ê°€
loss, accuracy = model.evaluate(X_test_padded, y_test, verbose=0)
print(f"\n**í…ŒìŠ¤íŠ¸ ë°ì´í„° ìµœì¢… í‰ê°€ ê²°ê³¼**")
print(f"Loss (ì†ì‹¤): {loss:.4f}")
print(f"Accuracy (ì •í™•ë„): {accuracy:.4f}")

# 4-3. í˜¼ë™ í–‰ë ¬(Confusion Matrix) ì‹œê°í™”
y_pred_probs = model.predict(X_test_padded)
# 0.5ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì´ì§„ ë¶„ë¥˜ (0 ë˜ëŠ” 1)
y_pred = (y_pred_probs > 0.5).astype(int)

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(7, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'], 
            yticklabels=['Actual 0', 'Actual 1'], 
            linecolor='black', linewidths=0.5)
plt.title('Confusion Matrix for Binary Classification', fontsize=15)
plt.ylabel('True Label (ì‹¤ì œ ë ˆì´ë¸”)', fontsize=12)
plt.xlabel('Predicted Label (ì˜ˆì¸¡ ë ˆì´ë¸”)', fontsize=12)
plt.show()

print("\n--- í˜¼ë™ í–‰ë ¬ ìƒì„¸ ê²°ê³¼ ---")
print(f"**ì‹¤ì œ 0ì„ 0ìœ¼ë¡œ ì˜ˆì¸¡ (ì •ë‹µ):** {cm[0, 0]} ê±´")
print(f"**ì‹¤ì œ 1ì„ 1ë¡œ ì˜ˆì¸¡ (ì •ë‹µ):** {cm[1, 1]} ê±´")
print(f"**ì‹¤ì œ 0ì„ 1ë¡œ ì˜ˆì¸¡ (ì˜¤ë¶„ë¥˜):** {cm[0, 1]} ê±´")
print(f"**ì‹¤ì œ 1ì„ 0ìœ¼ë¡œ ì˜ˆì¸¡ (ì˜¤ë¶„ë¥˜):** {cm[1, 0]} ê±´")
```

```python


# 1. Simple Dense (MLP) ëª¨ë¸ êµ¬ì¶•
# CNN ëª¨ë¸ê³¼ ë™ì¼í•œ Embedding ë ˆì´ì–´ë¥¼ ì‚¬ìš©í•˜ì—¬ ë™ì¼í•œ ì…ë ¥ ë²¡í„°ë¥¼ ë§Œë“­ë‹ˆë‹¤.
# Conv1D ëŒ€ì‹  Flattenì„ ì‚¬ìš©í•˜ì—¬ ì„ë² ë”© ë²¡í„°ë¥¼ í¼ì¹œ í›„ Dense ë ˆì´ì–´ì— ì…ë ¥í•©ë‹ˆë‹¤.

simple_model = Sequential([
    # CNNê³¼ ë™ì¼í•œ Embedding ë ˆì´ì–´
    Embedding(vocab_size, embedding_dim, input_length=maxlen),
    
    # Flatten ë ˆì´ì–´: [ë°°ì¹˜, ì‹œí€€ìŠ¤ ê¸¸ì´, ì„ë² ë”© ì°¨ì›] -> [ë°°ì¹˜, ì‹œí€€ìŠ¤ ê¸¸ì´ * ì„ë² ë”© ì°¨ì›]ìœ¼ë¡œ ë³€í™˜
    tf.keras.layers.Flatten(), 
    
    # ë“œë¡­ì•„ì›ƒ: ê³¼ì í•© ë°©ì§€
    Dropout(0.5),
    
    # ë°€ì§‘ ì—°ê²° ë ˆì´ì–´
    Dense(64, activation='relu'),
    
    # ì¶œë ¥ ë ˆì´ì–´: ì´ì§„ ë¶„ë¥˜ì´ë¯€ë¡œ sigmoid ì‚¬ìš©
    Dense(1, activation='sigmoid')
])

# ëª¨ë¸ ì»´íŒŒì¼
simple_model.compile(optimizer='adam',
                     loss='binary_crossentropy',
                     metrics=['accuracy'])

print("--- Simple Dense Model Summary ---")
simple_model.summary()

# 2. ëª¨ë¸ í•™ìŠµ
# ê¸°ì¡´ CNN ëª¨ë¸ê³¼ ë™ì¼í•œ ì—í¬í¬ ë° ë°°ì¹˜ í¬ê¸° ì‚¬ìš© ê¶Œì¥
print("\n--- Simple Dense Model í•™ìŠµ ì‹œì‘ ---")
simple_history = simple_model.fit(X_train_padded, y_train,
 epochs=10, 
 batch_size=32,
 validation_data=(X_val_padded, y_val), # ê²€ì¦ ë°ì´í„°(X_val) ì‚¬ìš©ìœ¼ë¡œ ìˆ˜ì •
 verbose=1)

print("\n--- Simple Dense Model í•™ìŠµ ì™„ë£Œ ---")
```

```python
# 1. í…ŒìŠ¤íŠ¸ ë°ì´í„° ìµœì¢… í‰ê°€ (Simple Dense)
loss_simple, acc_simple = simple_model.evaluate(X_test_padded, y_test, verbose=0)
print(f"**Simple Dense Layer Model ìµœì¢… ì •í™•ë„:** {acc_simple:.4f}")

# 2. CNN ëª¨ë¸ ìµœì¢… ì •í™•ë„ (ê¸°ì¡´ history ê°ì²´ì—ì„œ ê°€ì ¸ì˜¤ê¸°)
# (ì£¼ì˜: ê¸°ì¡´ history ê°ì²´ê°€ ì—†ìœ¼ë©´ ì´ ë¶€ë¶„ì€ ì£¼ì„ ì²˜ë¦¬í•˜ê±°ë‚˜ ìˆ˜ë™ìœ¼ë¡œ ì…ë ¥í•´ì•¼ í•©ë‹ˆë‹¤.)
# history.history['val_accuracy'][-1]ëŠ” CNNì˜ ìµœì¢… ê²€ì¦ ì •í™•ë„ë¥¼ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.
acc_cnn = history.history['val_accuracy'][-1]
print(f"**CNN Model ìµœì¢… ì •í™•ë„:** {acc_cnn:.4f}")


# 3. í•™ìŠµ ê³¡ì„  ì‹œê°í™” ë¹„êµ

plt.figure(figsize=(14, 6))

# ì •í™•ë„ ë¹„êµ
plt.subplot(1, 2, 1)
plt.plot(history.history['val_accuracy'], label='CNN Validation Accuracy')
plt.plot(simple_history.history['val_accuracy'], label='Simple Dense Validation Accuracy')
plt.title('Validation Accuracy Comparison', fontsize=15)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

# ì†ì‹¤(Loss) ë¹„êµ
plt.subplot(1, 2, 2)
plt.plot(history.history['val_loss'], label='CNN Validation Loss')
plt.plot(simple_history.history['val_loss'], label='Simple Dense Validation Loss')
plt.title('Validation Loss Comparison', fontsize=15)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# 4. ë§‰ëŒ€ ê·¸ë˜í”„ë¡œ ìµœì¢… ì •í™•ë„ ë¹„êµ
models = ['CNN Model', 'Simple Dense Model']
accuracies = [acc_cnn, acc_simple]

plt.figure(figsize=(6, 5))
bars = plt.bar(models, accuracies, color=['skyblue', 'lightcoral'])
plt.ylim(min(accuracies)*0.9, 1.0) # yì¶• ë²”ìœ„ ì¡°ì •
plt.title('Accuracy Comparison')
plt.ylabel('Accuracy')

# ë§‰ëŒ€ ìœ„ì— ì •í™•ë„ ê°’ í‘œì‹œ
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.4f}', ha='center', va='bottom')

plt.show()
```

ëª¨ë¸ ìœ í˜•: CNN (Convolutional Neural Network)

í•µì‹¬ ì•„ì´ë””ì–´: ë¬¸ì¥ì„ ì„ë² ë”©í•œ í›„ 1ì°¨ì› í•©ì„±ê³±(Conv1D) ìœ¼ë¡œ n-gram íŒ¨í„´(ì˜ˆ: ë‹¨ì–´ 3~5ê°œ ë¬¶ìŒ)ì„ ìë™ìœ¼ë¡œ ì¶”ì¶œ.

GlobalMaxPooling1D ë¡œ ë¬¸ì¥ì—ì„œ ê°€ì¥ ì¤‘ìš”í•œ íŠ¹ì§•ë§Œ ì¶”ë¦¼.

íŠ¹ì§• ì¶”ì¶œ ë°©ë²•: CNN ì»¤ë„ì´ ë‹¨ì–´ ìˆœì„œì™€ ì§€ì—­ ë¬¸ë§¥(local context) ì„ ë°˜ì˜í•´ ì˜ë¯¸ ìˆëŠ” íŒ¨í„´ì„ ê°ì§€.

ëª¨ë¸ ìœ í˜•: MLP (Multi-Layer Perceptron, ë‹¤ì¸µ í¼ì…‰íŠ¸ë¡ )

í•µì‹¬ ì•„ì´ë””ì–´: ë¬¸ì¥ì„ ë‹¨ì–´ ì„ë² ë”©(Embedding) ìœ¼ë¡œ ë°”ê¾¸ê³ , ì´ë¥¼ 1ì°¨ì› ë²¡í„°ë¡œ í‰íƒ„í™”(Flatten) í•œ ë’¤ ì™„ì „ì—°ê²°ì¸µ(Dense layer)ìœ¼ë¡œ ë¶„ë¥˜.
